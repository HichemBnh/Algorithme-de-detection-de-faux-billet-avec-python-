## Test du modle et analyse des rsidus

On analyse la distribution des rsidus avec un test de normalit et on calcul sa moyenne

```python
shapiro(reg_multi.resid)
```

```python
print('Moyenne des rsidus',reg_multi.resid.mean())
```

Le test nous indique que les rsidus ne suivent pas une loi normale, nanmoins la moyenne est trs proche de 0, on peut considrer quelle se rapproche dune loi normale 

On teste notre modle sur xtest et on rcupre les erreurs afin de les analyser 

```python
y_pred = reg_multi.predict(xtest)
df_error = pd.concat([y_pred,ytest], axis = 1)
df_error.columns= ['y_pred','ytest']
df_error['error'] =  df_error['ytest'] - df_error['y_pred'] 
df_error
```

```python
print("Mean Absolute Error MAE : ", mean_absolute_error(ytest,y_pred ))
print("Erreur Quadratique Moyenne RMSE : ", np.sqrt(mean_squared_error(ytest,y_pred)) )
print("Median Absolute Error : ", median_absolute_error(ytest,y_pred ))
```

En moyenne, les prdictions ont une erreur absolue de 0.3983 par rapport aux valeurs relles

La moyenne des erreurs quadratiques entre les valeurs prdites et les valeurs relles est de 0.5164, le RMSE est plus lev que la MAE car il est plus sensible aux Outliers
