## Création et entrainement du modèle

On effectue une régression pas à pas on supprimant les variables les moins significatives 

```python
reg_multi = smf.ols('margin_low~+margin_up+diagonal+length+height_left+height_right', data=train).fit()
print(reg_multi.summary())
```

On supprime l'intercept qui a une pvalue de 0.093

```python
reg_multi = smf.ols('margin_low~+margin_up+diagonal+length+height_left+height_right-1', data=train).fit()
print(reg_multi.summary())
```

On supprime diagonal qui a une pvalue de 0.298

```python
reg_multi = smf.ols('margin_low~+margin_up+length+height_left+height_right-1', data=train).fit()
print(reg_multi.summary())
```

Toutes les pvalues sont à 0, on peut garder ce modèle

On supprime de nos df train et test les colonnes qui n'ont pas été retenues dans ce modèle

```python
train = train[['height_left','height_right','margin_up','length','margin_low']]
print(train.shape)
train.head()
```

```python
xtest= xtest[['height_left','height_right','margin_up','length']]
print(xtest.shape)
xtest.head()
```

'ytest' pour vérrifier nos résultats

```python
print(ytest.shape)
ytest.head()
```

## Test du modèle et analyse des résidus

On analyse la distribution des résidus avec un test de normalité et on calcul sa moyenne

```python
shapiro(reg_multi.resid)
```

```python
print('Moyenne des résidus',reg_multi.resid.mean())
```

Le test nous indique que les résidus ne suivent pas une loi normale, néanmoins la moyenne est très proche de 0, on peut considérer qu’elle se rapproche d’une loi normale 

On teste notre modèle sur xtest et on récupère les erreurs afin de les analyser 

```python
y_pred = reg_multi.predict(xtest)
df_error = pd.concat([y_pred,ytest], axis = 1)
df_error.columns= ['y_pred','ytest']
df_error['error'] =  df_error['ytest'] - df_error['y_pred'] 
df_error
```

On vérifie le MAE et le RMSE  

```python
print("Mean Absolute Error MAE : ", mean_absolute_error(ytest,y_pred ))
print("Erreur Quadratique Moyenne RMSE : ", np.sqrt(mean_squared_error(ytest,y_pred)) )
print("Median Absolute Error : ", median_absolute_error(ytest,y_pred ))
```

En moyenne, les prédictions ont une erreur absolue de 0.3983 par rapport aux valeurs réelles

La moyenne des erreurs quadratiques entre les valeurs prédites et les valeurs réelles est de 0.5164, le RMSE est plus élevé que la MAE car il est plus sensible aux Outliers

## Partitionnement

Nous allons partitionner notre tableau en deux parties 80% pour le train et 20% pour les tests

```python
X = billets.drop('is_genuine', axis = 1)
y = billets[['is_genuine']]

x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=5)
```

```python
print('dimenssion de x_train :', x_train.shape)
print('dimenssion de y_train :', y_train.shape)
print('dimenssion de x_test :', x_test.shape)
print('dimenssion de y_test :', y_test.shape)
```

## Régression logistique avec Statmodel

```python
df_logi=pd.concat([x_train,y_train], axis = 1)
print(df_logi.shape)
df_logi.head()
```

On effectue une régression pas à pas en supprimant les variables les moins significatives 

```python
reg_log = smf.glm('is_genuine~margin_low+margin_up+diagonal+length+height_left+height_right',
                   data=df_logi, family=sm.families.Binomial()).fit()
print(reg_log.summary())
```

On supprime diagonal

```python
reg_log = smf.glm('is_genuine~margin_low+margin_up+length+height_left+height_right',
                   data=df_logi, family=sm.families.Binomial()).fit()
print(reg_log.summary())
```

On supprime height_left

```python
reg_log = smf.glm('is_genuine~margin_low+margin_up+length+height_right',
                   data=df_logi, family=sm.families.Binomial()).fit()
print(reg_log.summary())
```

On supprime height_right     

```python
reg_log = smf.glm('is_genuine~margin_low+margin_up+length',
                   data=df_logi, family=sm.families.Binomial()).fit()
print(reg_log.summary())
```

Model retenu

## colinéarité des variables

```python
variables = reg_log.model.exog
[variance_inflation_factor(variables, i) for i in np.arange(1,variables.shape[1])]
```

Toute nos valeurs sont inférieures à 10, il n’y a donc pas d'influence liée à la colinéarité des variables.

On supprime de notre df x_test les colonnes qui n’ont pas été retenues dans la création du modèle 

```python
x_test_logi =  x_test.drop(['height_left', 'diagonal', 'height_right'], axis = 1)
```

```python
x_test_logi.head()
```

## Test du modèle

On récupère les probabilités pour True

```python
y_pred_prob = abs(1-reg_log.predict(x_test_logi))
```

```python
y_pred_prob.head()
```

## visualisation sur la courbe de roc

```python
false_pos, true_pos, thresholds = roc_curve(y_test,y_pred_prob  )
```

```python
soc_score = roc_auc_score(y_test, y_pred_prob )
print('Air sous la courbe est de:' ,soc_score)
```

```python
plt.plot(false_pos, true_pos)
plt.plot([0, 1], [0, 1], linestyle='--')
plt.title('Courbe de ROC')
plt.xlabel('false_pos')
plt.ylabel('true_pos')
plt.show()
```

En calculant le meilleur seuil avec les vrais positifs et les faux positifs, nous obtenons : 

```python
print(thresholds[argmax(true_pos-false_pos)])
```

Ce seuil sera appliqué sur la probabilité pour minimiser les erreurs 

```python
best_threshold = thresholds[argmax(true_pos-false_pos)]
```

```python
y_pred_logi  = (y_pred_prob >=best_threshold).astype(int)
```

Nous créons un tableau avec les principaux indicateurs afin de comparer les résultats entre nos différents modèles de classification 

```python
df_reg_statmodels = rapport_score_regression(y_test, y_pred_logi, 'reg_statmodels')
df_reg_statmodels
```

```python
confusion_matrix_view(y_test, y_pred_logi)
```

Aux vues de nos indicateurs et de la matrice de confusion, notre modèle prédit correctement 99% des résultats. Il arrive à détecter 99.48 % des échantillons positifs, 98.13 % des échantillons faux sont classés correctement, et 98.97 % des échantillons positifs sont bien classés

## Méthode K-means

### Choix du nombre de clusters

```python
inertia = []
for k in range(1, 15):
    kmeans = KMeans(n_clusters = k, init = "random", n_init = 20).fit(x_train)
    inertia = inertia + [kmeans.inertia_]
inertia = pd.DataFrame({"k": range(1, 15), "inertia": inertia})
inertia.plot.line(x = "k", y = "inertia")
plt.scatter(2, inertia.query('k ==2')["inertia"], c = "red")
plt.show()
```

```python
# Une liste vide pour enregistrer les inerties :  
intertia_list = [ ]

# Notre liste de nombres de clusters : 
k_list = range(1, 10)

# Pour chaque nombre de clusters : 
for k in k_list : 
    
    # On instancie un k-means pour k clusters
    kmeans = KMeans(n_clusters=k)
    
    # On entraine
    kmeans.fit(x_train)
    
    # On enregistre l'inertie obtenue : 
    intertia_list.append(kmeans.inertia_)
for i in range(1,10):
    print(i, intertia_list[i-1] )
```
